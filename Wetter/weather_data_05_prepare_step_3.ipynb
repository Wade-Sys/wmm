```python
# Pickl-Dateipfade vorbereiten:
path_to_working_weather_files = '/home/paul/python_projects/masterthesis/data/wetter/'

pkl_file_wetter_step_2 = 'daten_wetter_step_2.p'
```

```python
# Pkl-Datei einlesen
df_wetter_pkl = pickle.load(open(path_to_working_weather_files + pkl_file_wetter_step_2, 'rb'))
df_wetter_pkl.head()
```

```python
# Dataframe umkopieren
df_wetter_1 = df_wetter_pkl.copy()
df_wetter_1.head()
```

```python
# Dateframe f체r die Aggregation vorbereiten  (nur # TMP_AT):
df_wetter_tmp_1 = df_wetter_1.copy() 
df_wetter_tmp_1.head()
```

```python
# Aggregation der Temperatur; Mean, Min, Max, Median
df_wetter_tmp_1_grp = df_wetter_tmp_1.groupby(['Jahr','Ort','Datum'], as_index=False, group_keys=False).agg({
    'TMP_AT' : ['mean','min','max','median']
}).copy()
```

```python
# Pr체fung der Anzahl der Aggregierten Datens채tze
df_wetter_tmp_1_grp.groupby(['Jahr']).count()
```

```python
# Pr체fung
df_wetter_tmp_1_grp.head()
```

```python
# Dataframe auf in flache Struktur bringen
df_wetter_tmp_1_grp.columns = ['_'.join(col) for col in df_wetter_tmp_1_grp.columns.values]
df_wetter_tmp_1_grp.head()
```

```python
# Spalten umbennen
df_wetter_tmp_1_grp.rename(columns={
    'Jahr_':'Jahr','Ort_':'Ort','Datum_':'Datum'
    ,'TMP_AT_mean':'TMP_MEAN','TMP_AT_min':'TMP_MIN','TMP_AT_max':'TMP_MAX','TMP_AT_median':'TMP_MEDIAN'}, inplace=True)
df_wetter_tmp_1_grp.head()
```

```python
# Dataframe umkopieren
df_wetter_2 = df_wetter_tmp_1_grp.copy()
df_wetter_2.head()
```

```python
# Dataframe pr체fen: F체r jeden Ort und Jahr muss ein aggregierter Wert vorhanden sein
df_wetter_2.groupby(['Jahr','Ort']).count()
```

```python
# Alle Datens채tze in eine CSV-Datei speichern. - Endg체ltige Dateie
file_name_prepared = path_to_working_weather_files + "daten_wetter_tmp.csv"
df_wetter_2.to_csv(file_name_prepared, sep=';', index=False, encoding='utf8')
```
