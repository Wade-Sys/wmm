```python
# Tabelle mit relevanten Wetterstationen erstellen
dict_wetter_stationen = {
    'ORT': ['Berlin','Berlin','Tokyo','Tokyo','Tokyo','London','NewYork','NewYork','Chicago'],    
    'STATION': ['10382099999','10385099999','47671099999','47662099999','47687099999','03768399999','72505394728','72502014734','72534014819']
}
df_wetter_stationen = pd.DataFrame(dict_wetter_stationen)
df_wetter_stationen.head()

```

```python
# Gemeinsames Dataframe für alle eingelesenen Daten aus der CSV erzeugen. (nur relevante Spalten)
df_wetter_daten_csv = pd.DataFrame(pd.DataFrame(
    columns=[
       'STATION','DATE','SOURCE','LATITUDE','LONGITUDE','ELEVATION','NAME','REPORT_TYPE','QUALITY_CONTROL','TMP' 
    ])
    ,dtype=str)
```

```python
# Pfad zu den Wetterdaten
path_to_downloaded_weather_files = '/home/paul/python_projects/masterthesis/data/wetter/downloaded/'
wetter_file_names = os.listdir(path_to_downloaded_weather_files)
```

```python
# Wetterdaten (CSV-Dateien) in Dataframe einlesen. (nur relevante Spalten)
for csvfile in wetter_file_names:
    if csvfile[-3:] == 'csv': # Nur CSV-Dateien
        wetter_file_path = path_to_downloaded_weather_files + csvfile
        print(wetter_file_path)
        df_wetter_daten_csv_temp = pd.read_csv(wetter_file_path,header=0, sep=',', decimal=',', quotechar='"',dtype=str
            ,usecols=['STATION','DATE','SOURCE','LATITUDE','LONGITUDE','ELEVATION','NAME','REPORT_TYPE','QUALITY_CONTROL','TMP' ]
        )
        df_wetter_daten_csv = df_wetter_daten_csv.append(df_wetter_daten_csv_temp)

df_wetter_daten_csv.head()
```

```python
# Ort zu den Stationen Joinen
df_wetter_daten_raw = pd.merge(left=df_wetter_stationen, right=df_wetter_daten_csv, on="STATION")
df_wetter_daten_raw.head()
```
