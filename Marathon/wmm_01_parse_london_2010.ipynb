```python
import time
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import numpy as np
import pandas as pd
```

<h1># 01 - Ergebnisdaten: Parsen der Ergebnisdaten - London (2010)</h1>
<hr>
<p><b>Hinweis:</b> Dateifpade sind absolut angegeben und m√ºssen entsprechend der eigenen Verzeichnisstruktur angepasst werden!</p>

```python
for year in range(2010,2011):
    # Leeres Dataframe anlegen
    df_results = pd.DataFrame(pd.DataFrame(
        columns=[
            'Jahr','Ges', 'Name','Club', 'Category', 'Runner_Number', 'Place_Gender', 'Place_Category','Place_Overall'
            ,'Start_Time_of_day','Finish_Time', 'Race_Status','Last_Split'
            ,'5K_Time_of_day','5K_Time','5K_Diff','5K_min_km','5K_kmh'
            ,'10K_Time_of_day','10K_Time','10K_Diff','10K_min_km','10K_kmh'
            ,'15K_Time_of_day','15K_Time','15K_Diff','15K_min_km','15K_kmh'
            ,'20K_Time_of_day','20K_Time','20K_Diff','20K_min_km','20K_kmh'
            ,'HK_Time_of_day','HK_Time','HK_Diff','HK_min_km','HK_kmh'
            ,'25K_Time_of_day','25K_Time','25K_Diff','25K_min_km','25K_kmh'
            ,'30K_Time_of_day','30K_Time','30K_Diff','30K_min_km','30K_kmh'
            ,'35K_Time_of_day','35K_Time','35K_Diff','35K_min_km','35K_kmh'
            ,'40K_Time_of_day','40K_Time','40K_Diff','40K_min_km','40K_kmh'
            ,'FK_Time_of_day','FK_Time','FK_Diff','FK_min_km','FK_kmh'
            ])
        ,dtype=str)
    
    # Treiber laden
    driver = webdriver.Firefox()
    wait = WebDriverWait(driver, 10)

    # Seite laden
    url = "https://results.virginmoneylondonmarathon.com/" + str(year) + "/"
    driver.get(url)
    time.sleep(3)

    for ges in ['M', 'W']:
        # Ergebnisse auswaehlen und anzeigen
        # Event
        if ges == 'M':
            event_ges = 'MAS'
        elif ges == 'W':
            event_ges = 'ELIW'            
        xpath_select_result = "//form[@id='form_search']//select[@name='event']//option[@value='" + event_ges + "']"
        driver.find_element_by_xpath(xpath_select_result).click()
        time.sleep(3)
        # Geschlecht
        xpath_select_result = "//form[@id='form_search']//select[@name='sex']//option[@value='" + ges + "']"
        driver.find_element_by_xpath(xpath_select_result).click()
        time.sleep(3)
        # Show Results Button
        xpath_select_result = "//form[@id='form_search']//input[@name='Submit']"
        driver.find_element_by_xpath(xpath_select_result).click()
        time.sleep(5)
        # Uber die ersten 10 Athleten iterieren
        for i in range(1,11):
            # Ein Athlet aufrufen        
            xpath_link = "//tbody[1]/tr[" + str(i) + "]/td[4]/a[1]"
            driver.find_element_by_xpath(xpath_link).click()
            time.sleep(2)
            # Auf der naechsten Seite die Daten abziehen
            xpath_table_names = "//div[@class='cbox-content']/div[@class='detail']/table[@class='list-table names']/tbody[1]"
            xpath_table_splits = "//div[@class='cbox-content']/div[@class='detail']/table[@class='list-table splits']/tbody[1]"
            xpath_table_total = "//div[@class='cbox-content']/div[@class='detail']/table[@class='list-table total']/tbody[1]"
            result_temp = {
                'Jahr' : year
                ,'Ges' : ges
                ,'Name' : driver.find_element_by_xpath(xpath_table_names + "/tr[1]/td[2]").text
                ,'Club' : ""
                ,'Category' : driver.find_element_by_xpath(xpath_table_names + "/tr[3]/td[2]").text
                ,'Runner_Number' : driver.find_element_by_xpath(xpath_table_names + "/tr[2]/td[2]").text 
                ,'Place_Gender' : driver.find_element_by_xpath(xpath_table_total + "/tr[2]/td[2]").text
                ,'Place_Category' : driver.find_element_by_xpath(xpath_table_total + "/tr[3]/td[2]").text
                ,'Place_Overall' : driver.find_element_by_xpath(xpath_table_total + "/tr[1]/td[2]").text
                ,'Start_Time_of_day' : driver.find_element_by_xpath(xpath_table_splits + "/tr[1]/td[2]").text
                ,'Finish_Time' : driver.find_element_by_xpath(xpath_table_total + "/tr[4]/td[2]").text
                ,'Race_Status' : ""
                ,'Last_Split' : ""
                ,'5K_Time_of_day' : ""
                ,'5K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[2]/td[2]").text
                ,'5K_Diff' : ""
                ,'5K_min_km' : ""
                ,'5K_kmh' : ""
                ,'10K_Time_of_day' : ""
                ,'10K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[3]/td[2]").text
                ,'10K_Diff' : ""
                ,'10K_min_km' : ""
                ,'10K_kmh' : ""
                ,'15K_Time_of_day' : ""
                ,'15K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[4]/td[2]").text
                ,'15K_Diff' : ""
                ,'15K_min_km' : ""
                ,'15K_kmh' : ""
                ,'20K_Time_of_day' : ""
                ,'20K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[5]/td[2]").text
                ,'20K_Diff' : ""
                ,'20K_min_km' : ""
                ,'20K_kmh' : ""
                ,'HK_Time_of_day' : ""
                ,'HK_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[6]/td[2]").text
                ,'HK_Diff' : ""
                ,'HK_min_km' : ""
                ,'HK_kmh' : ""
                ,'25K_Time_of_day' : ""
                ,'25K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[7]/td[2]").text
                ,'25K_Diff' : ""
                ,'25K_min_km' : ""
                ,'25K_kmh' : ""
                ,'30K_Time_of_day' : ""
                ,'30K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[8]/td[2]").text
                ,'30K_Diff' : ""
                ,'30K_min_km' : ""
                ,'30K_kmh' : ""
                ,'35K_Time_of_day' : ""
                ,'35K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[9]/td[2]").text
                ,'35K_Diff' : ""
                ,'35K_min_km' : ""
                ,'35K_kmh' : ""
                ,'40K_Time_of_day' : ""
                ,'40K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[10]/td[2]").text
                ,'40K_Diff' : ""
                ,'40K_min_km' : ""
                ,'40K_kmh' : ""
                ,'FK_Time_of_day' : ""
                ,'FK_Time' : ""
                ,'FK_Diff' : ""
                ,'FK_min_km' : ""
                ,'FK_kmh' : ""
            }
            df_results = df_results.append(result_temp, ignore_index=True)
            # Zurueck zu der Athletenliste
            driver.find_element_by_xpath("//div[@class='cbox-content']/div[@class='detail']/div[2]/a[1]").click()
            time.sleep(3)
        # Neue Suche
        driver.find_element_by_xpath("//div[@class='cbox-content']/div[@class='list']/div[@class='back']/a[1]").click()
        time.sleep(3)


    # Ergebnisse anzeigen
    #print(df_results.head())

    # Treiber schliessen
    driver.close()

    # Daten in die CSV Exportieren
    file_path = '/home/paul/python_projects/masterthesis/data/London/'
    file_name = 'daten_wmm_london_' + str(year) + '.csv'
    df_results.to_csv(file_path+file_name, sep=';', index=False)
```
