```python
# Treiber laden
driver = webdriver.Firefox()
driver.implicitly_wait(15)
#wait = WebDriverWait(driver, 10)

# Seite laden (Die erste Seite wird mit 2018 geladen)
url = "https://chicago-history.r.mikatiming.com/" + str(2018) + "/"
driver.get(url)
time.sleep(5)
```

```python
for year in range(2015, 2019):
    time.sleep(10)

    # Leeres Dataframe anlegen
    df_results = pd.DataFrame(pd.DataFrame(
    columns=[
        'Jahr','Ges', 'Name', 'Age_Group', 'Runner_Number', 'Place_Gender', 'Place_Age_Group','Place_Overall'
        ,'Start_Time_of_day','Finish_Time'
        ,'5K_Time','5K_Diff','5K_min_km','5K_kmh'
        ,'10K_Time','10K_Diff','10K_min_km','10K_kmh'
        ,'15K_Time','15K_Diff','15K_min_km','15K_kmh'
        ,'20K_Time','20K_Diff','20K_min_km','20K_kmh'
        ,'HK_Time','HK_Diff','HK_min_km','HK_kmh'
        ,'25K_Time','25K_Diff','25K_min_km','25K_kmh'
        ,'30K_Time','30K_Diff','30K_min_km','30K_kmh'
        ,'35K_Time','35K_Diff','35K_min_km','35K_kmh'
        ,'40K_Time','40K_Diff','40K_min_km','40K_kmh'
        ,'FK_Time','FK_Diff','FK_min_km','FK_kmh'
        ])
    ,dtype=str)

    for ges in ['M', 'W']:
        time.sleep(10)
        # Ergebnisse auswaehlen und anzeigen

        # Jahr
        xpath_select_result = "//select[@id='default-lists-event_main_group']//option[@value='" + str(year) + "']"
        driver.find_element_by_xpath(xpath_select_result).click()
        time.sleep(10)

        # Event
        xpath_select_result = "//select[@id='default-lists-event']//option[starts-with(@value,'MAR')]"
        driver.find_element_by_xpath(xpath_select_result).click()
        time.sleep(10)

        # Geschlecht
        xpath_select_result = "//select[@id='default-lists-sex']//option[@value='" + ges + "']"
        driver.find_element_by_xpath(xpath_select_result).click()
        time.sleep(10)

        # Show Results Button
        xpath_select_result = "//button[@id='default-submit']"
        driver.find_element_by_xpath(xpath_select_result).click()
        time.sleep(10)

        # Uber die ersten 10 Athleten iterieren (Header uebersprungen)
        for i in range(2,12):
            # Ein Athlet aufrufen        
            xpath_link = "//div[@id='cbox-main']/div[2]/ul[1]/li[" + str(i) + "]//h4[1]/a[1]"
            driver.find_element_by_xpath(xpath_link).click()
            time.sleep(3)

            # auf Kilometer umstellen
            driver.find_element_by_xpath("//button[@class='btn btn-default colgroup-switch-colgroup-splits colgroup-switch-colgroup-splits-metric']").click() 
            time.sleep(3)

            # Auf der naechsten Seite die Daten abziehen
            xpath_table_splits = "//div[@class='detail-box box-splits']//tbody[1]/"
            result_temp = {
                'Jahr' : driver.find_element_by_xpath("//td[@class='f-event_date last']").text
                ,'Ges' : ges
                ,'Name' : driver.find_element_by_xpath("//td[@class='f-__fullname last']").text
                ,'Age_Group' : driver.find_element_by_xpath("//td[@class='f-age_class last']").text
                ,'Runner_Number' : driver.find_element_by_xpath("//td[@class='f-start_no last']").text
                ,'Place_Gender' : driver.find_element_by_xpath("//td[@class='f-place_all last']").text
                ,'Place_Age_Group' : driver.find_element_by_xpath("//td[@class='f-place_age last']").text
                ,'Place_Overall' : driver.find_element_by_xpath("//td[@class='f-place_nosex last']").text
                ,'Start_Time_of_day' : driver.find_element_by_xpath("//td[@class='f-starttime_net last']").text
                ,'Finish_Time' : driver.find_element_by_xpath("//td[@class='f-time_finish_netto last']").text

                ,'5K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[1]/td[@class='time']").text
                ,'5K_Diff' : driver.find_element_by_xpath(xpath_table_splits + "/tr[1]/td[@class='diff']").text
                ,'5K_min_km' : driver.find_element_by_xpath(xpath_table_splits + "/tr[1]/td[@class='min_km right opt colgroup-splits colgroup-splits-metric']").text
                ,'5K_kmh' : driver.find_element_by_xpath(xpath_table_splits + "/tr[1]/td[@class='kmh colgroup-splits colgroup-splits-metric']").text

                ,'10K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[2]/td[@class='time']").text
                ,'10K_Diff' : driver.find_element_by_xpath(xpath_table_splits + "/tr[2]/td[@class='diff']").text
                ,'10K_min_km' : driver.find_element_by_xpath(xpath_table_splits + "/tr[2]/td[@class='min_km right opt colgroup-splits colgroup-splits-metric']").text
                ,'10K_kmh' : driver.find_element_by_xpath(xpath_table_splits + "/tr[2]/td[@class='kmh colgroup-splits colgroup-splits-metric']").text

                ,'15K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[3]/td[@class='time']").text
                ,'15K_Diff' : driver.find_element_by_xpath(xpath_table_splits + "/tr[3]/td[@class='diff']").text
                ,'15K_min_km' : driver.find_element_by_xpath(xpath_table_splits + "/tr[3]/td[@class='min_km right opt colgroup-splits colgroup-splits-metric']").text
                ,'15K_kmh' : driver.find_element_by_xpath(xpath_table_splits + "/tr[3]/td[@class='kmh colgroup-splits colgroup-splits-metric']").text

                ,'20K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[4]/td[@class='time']").text
                ,'20K_Diff' : driver.find_element_by_xpath(xpath_table_splits + "/tr[4]/td[@class='diff']").text
                ,'20K_min_km' : driver.find_element_by_xpath(xpath_table_splits + "/tr[4]/td[@class='min_km right opt colgroup-splits colgroup-splits-metric']").text
                ,'20K_kmh' : driver.find_element_by_xpath(xpath_table_splits + "/tr[4]/td[@class='kmh colgroup-splits colgroup-splits-metric']").text

                ,'HK_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[5]/td[@class='time']").text
                ,'HK_Diff' : driver.find_element_by_xpath(xpath_table_splits + "/tr[5]/td[@class='diff']").text
                ,'HK_min_km' : driver.find_element_by_xpath(xpath_table_splits + "/tr[5]/td[@class='min_km right opt colgroup-splits colgroup-splits-metric']").text
                ,'HK_kmh' : driver.find_element_by_xpath(xpath_table_splits + "/tr[5]/td[@class='kmh colgroup-splits colgroup-splits-metric']").text

                ,'25K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[6]/td[@class='time']").text
                ,'25K_Diff' : driver.find_element_by_xpath(xpath_table_splits + "/tr[6]/td[@class='diff']").text
                ,'25K_min_km' : driver.find_element_by_xpath(xpath_table_splits + "/tr[6]/td[@class='min_km right opt colgroup-splits colgroup-splits-metric']").text
                ,'25K_kmh' : driver.find_element_by_xpath(xpath_table_splits + "/tr[6]/td[@class='kmh colgroup-splits colgroup-splits-metric']").text

                ,'30K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[7]/td[@class='time']").text
                ,'30K_Diff' : driver.find_element_by_xpath(xpath_table_splits + "/tr[7]/td[@class='diff']").text
                ,'30K_min_km' : driver.find_element_by_xpath(xpath_table_splits + "/tr[7]/td[@class='min_km right opt colgroup-splits colgroup-splits-metric']").text
                ,'30K_kmh' : driver.find_element_by_xpath(xpath_table_splits + "/tr[7]/td[@class='kmh colgroup-splits colgroup-splits-metric']").text

                ,'35K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[8]/td[@class='time']").text
                ,'35K_Diff' : driver.find_element_by_xpath(xpath_table_splits + "/tr[8]/td[@class='diff']").text
                ,'35K_min_km' : driver.find_element_by_xpath(xpath_table_splits + "/tr[8]/td[@class='min_km right opt colgroup-splits colgroup-splits-metric']").text
                ,'35K_kmh' : driver.find_element_by_xpath(xpath_table_splits + "/tr[8]/td[@class='kmh colgroup-splits colgroup-splits-metric']").text

                ,'40K_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[9]/td[@class='time']").text
                ,'40K_Diff' : driver.find_element_by_xpath(xpath_table_splits + "/tr[9]/td[@class='diff']").text
                ,'40K_min_km' : driver.find_element_by_xpath(xpath_table_splits + "/tr[9]/td[@class='min_km right opt colgroup-splits colgroup-splits-metric']").text
                ,'40K_kmh' : driver.find_element_by_xpath(xpath_table_splits + "/tr[9]/td[@class='kmh colgroup-splits colgroup-splits-metric']").text

                ,'FK_Time' : driver.find_element_by_xpath(xpath_table_splits + "/tr[10]/td[@class='time']").text
                ,'FK_Diff' : driver.find_element_by_xpath(xpath_table_splits + "/tr[10]/td[@class='diff']").text
                ,'FK_min_km' : driver.find_element_by_xpath(xpath_table_splits + "/tr[10]/td[@class='min_km right opt colgroup-splits colgroup-splits-metric']").text
                ,'FK_kmh' : driver.find_element_by_xpath(xpath_table_splits + "/tr[10]/td[@class='kmh colgroup-splits colgroup-splits-metric']").text
            }
            df_results = df_results.append(result_temp, ignore_index=True)
            # Zurueck zu der Athletenliste
            driver.find_element_by_xpath("//div[@class='row buttons']//div[@class='btn-group']/a[1]").click()
            time.sleep(3)

        # Neue Suche
        driver.find_element_by_xpath("//div[@id='cbox-main']/div[3]//div[@class='btn-group']/a[1]").click()
        time.sleep(3)
    
    # Ergebnisse anzeigen
    # print(df_results.head())

    # Daten in die CSV Exportieren
    file_path = '/home/paul/python_projects/masterthesis/data/Chicago/'
    file_name = 'daten_wmm_chicago_' + str(year) + '.csv'
    df_results.to_csv(file_path+file_name, sep=';', index=False)
```
